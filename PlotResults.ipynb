# Plot Each Metric in Separate Figure

```python
import glob
import pandas as pd
import os
import matplotlib.pyplot as plt

# set the path where the CSV files are stored
#load_path = './MNISTRunPy/MNIST_CSV_Aggregated/'
#save_fig_path = './MNISTRunPy/MNIST_Plots/'
load_path = './FashionMNISTRunPy/FashionMNIST_ResNet18_CSV_Aggregated/'
save_fig_path = './FashionMNISTRunPy/FashionMNIST_ResNet18_Plots/'

beta= ['0.01', '0.1', '0.3', '0.6', '0.9', '1','1.0']
Metrics_Y=['train_loss','train_acc','test_loss','test_acc'] 
Coef= 1

def myplot(metric_Y, load_path, save_fig_path, Coef):
    isExist = os.path.exists(save_fig_path)
    if not isExist:
        os.makedirs(save_fig_path)
    # create a list of all CSV files in the folder
    csv_files = glob.glob(load_path + '*.csv')
    # loop over the CSV files and group them by hyperparameters
    for csv_file in csv_files:
        # extract hyperparameters from file name
        SGDM_file_name = os.path.basename(csv_file)
        if '_SGD_' in SGDM_file_name:
            continue
        SGD_file_name = SGDM_file_name
        SGD_file_name = SGD_file_name.replace('SGDM','SGD')
        for i in beta:
            SGD_file_name = SGD_file_name.replace('_beta_'+i,'')
        SGDM = pd.read_csv(load_path+SGDM_file_name)
        SGD = pd.read_csv(load_path+SGD_file_name)
        plt.figure(figsize=[10, 5], dpi=72)
        ax = plt.gca()
        SGDM.plot(x='Unnamed: 0', y=metric_Y+'_mean',ax=ax,label='SGDM', linewidth=3)
        plt.fill_between(SGDM['Unnamed: 0'], SGDM[metric_Y+'_mean'] - Coef*SGDM[metric_Y+'_std'], SGDM[metric_Y+'_mean'] + Coef*SGDM[metric_Y+'_std'], alpha=0.3)
        SGD.plot(x='Unnamed: 0', y=metric_Y+'_mean',ax=ax, label='SGD', linewidth=3)
        plt.fill_between(SGD['Unnamed: 0'], SGD[metric_Y+'_mean'] - Coef*SGD[metric_Y+'_std'], SGD[metric_Y+'_mean'] + Coef*SGD[metric_Y+'_std'], alpha=0.3)
        plt.title(SGDM_file_name.split('.csv')[0])
        plt.ylabel(metric_Y)
        plt.xlabel('Epoch')
        plt.rcParams.update({'font.size': 15})
        plt.savefig(save_fig_path+SGDM_file_name.split('.csv')[0]+'_'+metric_Y+'.png', dpi=144, format=None, metadata=None, bbox_inches=None, pad_inches=0.1,
                facecolor='auto', edgecolor='auto', backend=None)
        print("The figures saved as: "+save_fig_path+SGDM_file_name.split('.csv')[0]+'_'+metric_Y+'.png')
    return

for metric_Y in Metrics_Y:
    myplot(metric_Y, load_path, save_fig_path,Coef)
```
